```r
# Load required packages
library(dplyr)
library(readr)
#library(scale)

# Load the nutrient data
# Load in our choice and demographic data

data=read.csv("https://raw.githubusercontent.com/CaitlinLloyd/Psychology_Programming2025/refs/heads/main/Data/demo.csv")
choices=read.csv("https://raw.githubusercontent.com/CaitlinLloyd/Psychology_Programming2025/refs/heads/main/Data/FCT_df.csv")
```

```r
colnames(data)
colnames(choices)
df_merged<- merge(data,choices,by="ID")
#df_merged
```

```r
df_merged
```

```r
df_merged%>%colnames()
```

```r
# Group by participant ID, block, fat, and sugar; exclude rows with missing data
summary_df <- df_merged %>%
  filter(!is.na(rating), !is.na(reaction_time)) %>%
  group_by(ID, block, fat, sugar) %>%
  summarise(
    average_rating = mean(rating, na.rm = TRUE),
    average_reaction_time = mean(reaction_time, na.rm = TRUE),
    trial_count = n(),  # count of non-missing trials
    .groups = 'drop'
  )

# View the result
summary_df
```

```r
#mean reaction time, high fat, sugar, correlation
# Step 3: Pivot wider on combinations of fat/sugar/kcal levels
library(tidyr)

summary_df$fat <- ifelse(summary_df$fat=="high","highfat","lowfat")
summary_df$sugar <- ifelse(summary_df$sugar=="high","highsugar","lowsugar")

df_wide <- summary_df %>%
  pivot_wider(
    names_from = c(block,fat,sugar),  ###if we do not have the block, every participants may have 3 block
    values_from = c(average_rating, average_reaction_time, trial_count)
  )

# Try again but labeling the food types so you can more clearly see what the column names align with
df_wide
```

```r
summary_df_demo <-merge(df_wide,data, by="ID")
```

```r
colnames(summary_df_demo)
```

```r
library(purrr)
library(broom)

# List of variables to test
vars_to_test <- c("BMI")

# Split groups
patients <- summary_df_demo %>% filter(Patient == "Yes")
controls <- summary_df_demo %>% filter(Patient == "No")

# Function to run Welch t-test on a variable, what did function do: takes argument and do in it
run_t_test <- function(var) {
  test_result <- t.test(patients[[var]], controls[[var]], var.equal = FALSE)
  tibble(
    variable = var,
    mean_patient = mean(patients[[var]], na.rm = TRUE),
    mean_control = mean(controls[[var]], na.rm = TRUE),
    t_stat = test_result$statistic,
    p_value = test_result$p.value
  )
}

# Apply to each variable, another way to do for loop
ttest_results <- map_dfr(vars_to_test, run_t_test)

# View results
print(ttest_results)
```

```r
## Load in the delay discounting data
data<-read.csv("https://raw.githubusercontent.com/CaitlinLloyd/Psychology_Programming2025/refs/heads/main/Data/DD_df.csv")
data
```

```r
##how many delayed choice per person in the control condition
control_trial <- subset(data, data$condition=="control")
count <- control_trial %>% group_by(participant_id, chose_delayed)%>% count()

count <- subset (count, chose_delayed==1)
demo<- data[,c("participant_id", "group", "age", "impulsivity")]
df<- merge(count, demo)
df<-df[!duplicated(df),]
```

```r
df
```

```r
## Calculate the average number of delayed choices in patient and HC groups
#patient <- subset(df, df$group=="SUD")
#mean(patient$n)

df %>%

  group_by(group) %>%
  summarise(
    average_numbers_delay = mean(n, na.rm = TRUE),
    .groups = 'drop'
  )
```

```r
library(dplyr)
library(tidyr)
# Calculate the correlation between age, delayed choices and impulsivity scale per group
control_trial <- subset(data, data$condition == "control")
pivot_table <- control_trial %>%
  group_by(participant_id, chose_delayed) %>%
  summarise(trial_count = n())
#pivot_table

pivot_wide <- pivot_table %>%
  pivot_wider(
    names_from = c(chose_delayed),
    values_from = c(trial_count),
  )
#pivot_wide

df1<-merge(pivot_wide,demo)
df1<-df1[!duplicated(df1),]

df1 <- df1 %>%
  rename(
    num_nondelayed = `0`,
    num_delayed = `1`
  )
df1
```

```r
df1 %>%
  group_by(group) %>%
  summarise(
    cor_age_delay = cor(age, num_delayed, use = "complete.obs"),
    cor_impulsivity_delay = cor(impulsivity, num_delayed, use = "complete.obs"),
    cor_age_impulsivity = cor(age, impulsivity, use = "complete.obs")
  )
```

```r
###Plot delayed choices and impulsivity by patient group
library(ggplot2)

ggplot(df1, aes(x = num_delayed, y = impulsivity,group=group,color=group)) +
  geom_point(size = 3, alpha = 0.8) +
  geom_smooth(method = "lm", se = TRUE, size = 1) +
  labs(
    title = "delayed choices and impulsivity",
    x = "Delayed Choice",
    y = "Impulsivity"
  ) +
  theme_minimal(base_size = 14) +
  theme(
    plot.title = element_text(face = "bold", hjust = 0.5, color = "pink")
  )

```
